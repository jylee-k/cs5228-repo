## CW2, Part 1: Regression and Classification (3+3= 6 marks)

### CW2-1: Regression (3 marks)
Datasets: cs5228-Housing.csv 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler

# 1. Load dataset
df = pd.read_csv('cs5228-Housing.csv')

# 3. Preprocessing: categorical to numerical
# Binary columns: yes=1, no=2 (avoiding zero as per instructions)
binary_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in binary_cols:
    df[col] = df[col].map({'yes': 1, 'no': 2})

# Furnishing status: avoid coding from zero
df['furnishingstatus'] = df['furnishingstatus'].map({'unfurnished': 1, 'semi-furnished': 2, 'furnished': 3})

# Define independent (X) and dependent (y) variables
# Features: area, bedrooms, bathrooms, stories, mainroad, guestroom, basement, hot water heating, air conditioning, parking, prefarea, furnishing status
X = df.drop(['price', 'class_label'], axis=1)
y = df['price']

# 2. Correlation analysis
correlation = X.corrwith(y)
print("Correlation analysis with price:\n", correlation)

# Variable removal justification:
# All features show reasonable correlation or theoretical importance. No features were discarded.

# 8, 11. Split 90/10 with random_state=110
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=110)

# 9. Hyperparameter optimization check
# LinearRegression in sklearn doesn't have significant hyperparameters to optimize for this task.

# 5, 7. Regression with and without normalization
def run_regression(X_train_data, X_test_data):
    model = LinearRegression()
    model.fit(X_train_data, y_train)
    y_pred_train = model.predict(X_train_data)
    y_pred_test = model.predict(X_test_data)
    
    mse_train = mean_squared_error(y_train, y_pred_train)
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mse_test = mean_squared_error(y_test, y_pred_test)
    mae_test = mean_absolute_error(y_test, y_pred_test)
    
    return mse_train, mae_train, mse_test, mae_test

# Without normalization
res_no_norm = run_regression(X_train, X_test)

# With normalization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
res_norm = run_regression(X_train_scaled, X_test_scaled)
```

```python
# Deliverable reporting
print("--- Deliverables for CW2-1 ---")
print("Discarded variables: None. All features are retained to capture full variance.")
print("\nFirst few rows of dataset:")
print(df.head())
print("\nDataset description:")
print(df.describe())
print(f"\nSize of training set: {X_train.shape[0]}")
print(f"Size of testing set: {X_test.shape[0]}")

print("\nResults without normalization:")
print(f"Train MSE: {res_no_norm[0]:.2f}, MAE: {res_no_norm[1]:.2f}")
print(f"Test MSE: {res_no_norm[2]:.2f}, MAE: {res_no_norm[3]:.2f}")

print("\nResults with normalization:")
print(f"Train MSE: {res_norm[0]:.2f}, MAE: {res_norm[1]:.2f}")
print(f"Test MSE: {res_norm[2]:.2f}, MAE: {res_norm[3]:.2f}")

print("\n--- Best results printed below (Normalization does not affect LinearRegression results) ---")
print(f"Final Test MSE: {res_no_norm[2]:.2f}")
print(f"Final Test MAE: {res_no_norm[3]:.2f}")
```

### CW2-2: Classification (3 marks) 
Datasets: cs5228-Housing.csv  

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score, confusion_matrix

X_clf = df.drop(['price', 'class_label'], axis=1)
y_clf = df['class_label']

# Split 90/10 with random_state=110
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.1, random_state=110)

# Define metrics function
def report_metrics(model, X_tr, y_tr, X_te, y_te):
    y_pred_tr = model.predict(X_tr)
    y_pred_te = model.predict(X_te)
    
    acc_tr = accuracy_score(y_tr, y_pred_tr)
    pr_tr = precision_score(y_tr, y_pred_tr, average='macro', zero_division=0)
    rc_tr = recall_score(y_tr, y_pred_tr, average='macro', zero_division=0)
    
    acc_te = accuracy_score(y_te, y_pred_te)
    pr_te = precision_score(y_te, y_pred_te, average='macro', zero_division=0)
    rc_te = recall_score(y_te, y_pred_te, average='macro', zero_division=0)
    
    return (acc_tr, pr_tr, rc_tr), (acc_te, pr_te, rc_te)

# Decision Tree
dt = DecisionTreeClassifier(random_state=110, max_depth=5)
dt.fit(X_train_clf, y_train_clf)
dt_res_tr, dt_res_te = report_metrics(dt, X_train_clf, y_train_clf, X_test_clf, y_test_clf)

# Random Forest (with simple optimization)
rf = RandomForestClassifier(random_state=110, n_estimators=200, max_depth=10, class_weight='balanced')
rf.fit(X_train_clf, y_train_clf)
rf_res_tr, rf_res_te = report_metrics(rf, X_train_clf, y_train_clf, X_test_clf, y_test_clf)

# Confusion Matrix labels order
labels = ['cheap', 'medium', 'expensive', 'very expensive']
cm = confusion_matrix(y_test_clf, rf.predict(X_test_clf), labels=labels)
```

```python
print("--- Deliverables for CW2-2 ---")
print("Discarded variables: None.")
print(f"\nSize of training set: {X_train_clf.shape[0]}")
print(f"Size of testing set: {X_test_clf.shape[0]}")

print("\nDecision Tree Results:")
print(f"Train Accuracy: {dt_res_tr[0]:.4f}, Precision: {dt_res_tr[1]:.4f}, Recall: {dt_res_tr[2]:.4f}")
print(f"Test Accuracy: {dt_res_te[0]:.4f}, Precision: {dt_res_te[1]:.4f}, Recall: {dt_res_te[2]:.4f}")

print("\nRandom Forest Results (Best Model):")
print(f"Train Accuracy: {rf_res_tr[0]:.4f}, Precision: {rf_res_tr[1]:.4f}, Recall: {rf_res_tr[2]:.4f}")
print(f"Test Accuracy: {rf_res_te[0]:.4f}, Precision: {rf_res_te[1]:.4f}, Recall: {rf_res_te[2]:.4f}")

print("\nConfusion Matrix (Random Forest, labels=['cheap', 'medium', 'expensive', 'very expensive']):")
print(cm)
```
