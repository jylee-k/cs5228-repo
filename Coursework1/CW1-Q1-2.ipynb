```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
# load the clean dataset from previous task
df = pd.read_csv('result1-1.csv')
print(f"dataset shape: {df.shape}")
```


For columns B (workclass), D (education), J (sex):

We use `pd.factorize` to get unique integer labels with label encoding.

```python
cols_to_encode = ['workclass', 'education', 'sex']

for col in cols_to_encode:
    df[col], _ = pd.factorize(df[col])

# Show first 15 records
print("first 15 records after encoding:")
df.head(15)
```

```python
# save result
df.to_csv('result1-2.csv', index=False)
```

Next, normalise the columns K and L using Z-transform.

```python
cols_to_norm = ['capital-gain', 'capital-loss']

# calculate mean and std before norm
print("Statistics BEFORE normalisation:")
print(df[cols_to_norm].agg(['mean', 'std']))
```

```python
# init scaler
scaler = StandardScaler()
df[cols_to_norm] = scaler.fit_transform(df[cols_to_norm])

# calculate mean and std after normalization
print("Statistics AFTER normalisation:")
print(df[cols_to_norm].agg(['mean', 'std']))
```

```python

# show last 15 records
print("last 15 records after normalization:")
df.tail(15)
```
