## Replacing missing value with average or ...

```python
# Display the original DataFrame
print("Original DataFrame:\n", df)

# Step 1: Calculate the mean of a column 
mean1 = df['weight'].mean()

# Step 2: Replace missing values in a column with the mean
df['weight'] = df['weight'].fillna(mean1)

# Display the modified DataFrame
print("\nDataFrame after replacing missing values with the mean in column:\n", df)
```

## Data Normalization 
#### I would like to normalize one of the numerical columns of my dataframe. 
#### I want to make it standard Gaussian, with mean=0 and std=1 (Z-Transform)

```python
# Display the original DataFrame
print("Original DataFrame:\n", df2)
df2 = df2.dropna()

# Choose the column to standardize, e.g., column 'height'
column_to_normalize = 'height'

# Step 1: Calculate the mean and standard deviation of the column
mean = df2[column_to_normalize].mean()
std_dev = df2[column_to_normalize].std()

# Step 2: Apply the standardization formula
df2[column_to_normalize + '_standardized'] = (df[column_to_normalize] - mean) / std_dev

# Display the DataFrame with the standardized column
print("\nDataFrame after standardizing column:\n", df2)
```

```python
column_to_normalize = 'weight'
mean = df2[column_to_normalize].mean()
std_dev = df2[column_to_normalize].std()
df2[column_to_normalize] = (df[column_to_normalize] - mean) / std_dev

print("\nDataFrame after standardizing column:\n", df2)
```
